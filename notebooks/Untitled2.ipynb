```python
import json
from decimal import Decimal
import numpy as np
import os, sys
from decimal import Decimal
sys.path.append(os.path.join(os.path.abspath('..'), 'src'))
from DocumentLinker import DocumentLinker
from datawrapper import DataWrapper
import json

# Determines percentage correct per type 
def maked3(vectorizer, distance):
    json_data= open('../data/final_analysis/{0}_{1}.json'.format(vectorizer, distance)).read()
    data = json.loads(json_data)
    result = []
    for doc in data:
        curResult = {'name': '{0}.{1}'.format(data[doc]['type'].lower(), data[doc]['title'].strip().strip('.')), 'size': 1000, 'imports': []}
        for link in data[doc]['links']:
            thisLink = data[doc]['links'][link]
            title = (thisLink['title']).strip('.')
            print title
            curResult['imports'].append('{0}.{1}'.format(thisLink['type'].lower(), title))
        result.append(curResult)
    file = open('d3.json', "w")
    file.write(json.dumps(result))
    file.close()
    print('File saved')
    
maked3('textvectorizer', 'cosine')
```
