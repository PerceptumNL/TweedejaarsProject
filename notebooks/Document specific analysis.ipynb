```python
import os, sys
import cPickle as pickle
from decimal import Decimal
sys.path.append(os.path.join(os.path.abspath('..'), 'src'))
from DocumentLinker import DocumentLinker
from datawrapper import DataWrapper
import operator
import pandas
import numpy as np
import scipy as sp

name_dict = pickle.load(open('../data/title_id_mapping.pickle'))

def getAverageDict(vectors, name_dict):
    vectors = [x[1].todense() for x in vectors]
    average = vectors[0]
    for i in vectors:
        average += i
    average = np.array(average/len(vectors))[0]
    average = dict((name_dict[i], average[i]) for i in range(0, len(average)))
    return average

def getVectors(new_doc, vectorList, data, vectorizer):
    vectorizer = getattr(__import__('vectorizers.' + vectorizer), vectorizer)
    data_bows, new_doc_bow, vec = vectorizer.vectorize(data, new_doc, True)
    name_dict = vec.get_feature_names()
    print vec.get_stop_words()
    average = getAverageDict(data_bows, name_dict)
    
    vectors = [(-1, new_doc_bow)]
    
    vectorDicts = []
    for i in vectorList:
        for x in data_bows:
            if x[0] == i:
                vectors.append(x)
                break
    for vector in vectors:
        indices = vector[1].indices
        vectorDict = {}
        for j in range (0, len(indices)):
            vectorDict[name_dict[indices[j]]] = vector[1].data[j]
        vectorDict = sorted(vectorDict.iteritems(), key=operator.itemgetter(1))
        vectorDicts.append((vector[0], vectorDict))
    return average, vectorDicts

def getCosineDiff(vector1, vector2, average):
    dict1 = {}
    dict2 = {}
    dict3 = {}

    for i, j in vector1:
        dict1[i] = j
    for i, j in vector2:
        dict2[i] = j
        if dict1.has_key(i):
            dict3[i] = abs(dict1[i] - j) - average[i]

    vectorDict1 = sorted(dict1.iteritems(), key=operator.itemgetter(1))
    vectorDict2 = sorted(dict2.iteritems(), key=operator.itemgetter(1))
    vectorDict3 = sorted(dict3.iteritems(), key=operator.itemgetter(1))

    return [vectorDict1, vectorDict2, vectorDict3]

# Get differences all real links of a document
def proposedLinksTable(item_id, vectorizer = 'textvectorizer', distance='cosine', k = 10):
    data = DataWrapper('../data/expert_maybe_true.pickle')
    new_doc, data = data.remove_item(item_id)
    linker = DocumentLinker(data, k)
    linker.get_links(new_doc, vectorizer, distance)
    
    nlinks = []
    for link in linker.links:
        nlinks.append(link[0])
    average, proposed = getVectors(new_doc, nlinks, data, vectorizer)
    
    # Set the item itself in proper format in the list
    diff = [getCosineDiff(proposed[0][1], proposed[0][1], average)[1]]
    
    # Now find differences for all other items
    for i in proposed:
        if i[0] == -1:
            continue
        diff.append(getCosineDiff(proposed[0][1], i[1], average)[2])

    df = pandas.DataFrame(diff).transpose()
    df.columns = [data.value_for_keys_with_item(new_doc, 'title', 'name')] + [data.value_for_keys(x, 'title', 'name') for x in nlinks]
    pandas.option_context('display.max_rows', 1000)
    return df
    

def trueLinksTable(item_id, vectorizer = 'textvectorizer', distance='cosine'):
    data = DataWrapper('../data/expert_maybe_true.pickle')
    new_doc, data = data.remove_item(item_id)
    linker = DocumentLinker(data, len(data.data['items']))
    linker.get_links(new_doc, vectorizer, distance)
    
    nlinks = []
    for link in linker.links:
        if link[0] in new_doc['links']:
            nlinks.append(link[0])
    average, proposed = getVectors(new_doc, nlinks, data, vectorizer)
    
    # Set the item itself in proper format in the list
    diff = [getCosineDiff(proposed[0][1], proposed[0][1], average)[1]]
    
    # Now find differences for all other items
    for i in proposed:
        if i[0] == -1:
            continue
        diff.append(getCosineDiff(proposed[0][1], i[1], average)[2])

    df = pandas.DataFrame(diff).transpose()
    df.columns = [0] + [data.value_for_keys(x, 'title', 'name') for x in nlinks]
    pandas.option_context('display.max_rows', 1000)
    return df

def table(item_id, k = 10):
    data = DataWrapper('../data/expert_maybe_true.pickle')
    new_doc, data = data.remove_item(11)
    linker = DocumentLinker(data, 10)
    linker.get_links(new_doc, 'textvectorizer', 'cosine')
    links = [x[0] for x in linker.links]
    links.append(5)
    average, proposed = getVectors(new_doc, links, data, vectorizer)

    df = pandas.DataFrame(map(lambda x: x[1], proposed)).transpose()
    df.columns = [0] + [data.value_for_keys(x, 'title', 'name') for x in links]
    pandas.option_context('display.max_rows', 231)
    df
```
