## Getting names to represent items
Apparently, not all items have the same properties (i.e. a document has a 'title' but a person has a 'name'). Thus to represent them properly we need to have a function that returns a recognizable string for each document type.

```python
def printRelationships(range = [0,100]):
    for item in data['items'].values()[range[0]:range[1]]:
        #print item
        print itemToShortString(item)
        for link in item['links']:
            #print link
            print("\t - {0}".format(itemToShortString(data['items'][link])))

def itemToShortString(item):
    if item.has_key('title'):
        return item['title']
    if item['type'] == 'Person':
        return item['name']
    return 'undefined ' + item['type'] 
    
```

```python
printRelationships([0,10])
```
